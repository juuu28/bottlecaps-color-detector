```python
from ultralytics import YOLO
import os
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset_yaml = """
path: relabel_dataset

train: images
val: images

nc: 3

names:
  0: light_blue
  1: dark_blue
  2: others
"""

with open("dataset.yaml", "w") as f:
    f.write(dataset_yaml)

print(open("dataset.yaml").read())
```

```python
DATASET_YAML = "dataset.yaml"  # update path if needed

assert os.path.exists(DATASET_YAML), "dataset.yaml not found!"
print(open(DATASET_YAML).read())
```

```python
import wandb
wandb.login()
```

```python
MODELS_TO_TEST_V8 = [
    "yolov8n.pt",
    "yolov8s.pt",
    "yolov8m.pt",  
]
RESULTS_V8 = []

wandb.init(
    project="bottlecaps-color-detection",  
    name="yolov8-multi"         
)

for model_name in MODELS_TO_TEST_V8:
    print(f"\n=== Training {model_name} ===")

    model = YOLO(model_name)

    results = model.train(
        data=DATASET_YAML,
        epochs=20,
        imgsz=640,
        batch=16,
        name=f"exp_{model_name.replace('.pt','')}",
        patience=10,
        project="bottlecaps-color-detection",   # wandb project name
        exist_ok=True                    # avoid error if folder exists
    )

    RESULTS_V8.append({
        "model": model_name,
        "results_dir": results.save_dir,
    })
```
